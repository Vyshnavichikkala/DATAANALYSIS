```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
 url = "https://en.wikipedia.org/wiki/List_of_Miss_Universe_titleholders"
```

```python
page = requests.get(url)
```

```python
soup = BeautifulSoup(page.text,'html')
print(soup)
```

```python
table = soup.find('table')
print(table)
```

```python
Headings = table.find_all('th')
print(Headings)
```

```python
Header = [i.text for i in Headings]
print(Header)
```

```python
df = pd.DataFrame(columns = Header)
df
```

```python
rows = table.find_all('tr')
print(rows)
```

```python
ind=0
for i in rows[1:6]:
  row_data = i.find_all('td')
  every_row_data = [i.text for i in row_data]
  print(every_row_data)
df.loc[ind] = every_row_data[:len(df.columns)]
ind = ind+1
```

```python
df
```

```python
df.to_csv("universe.csv")
```

```python
p = requests.get(url)
```

```python
t = code.find('table')
print(t)
```

```python
data = t.find_all('td')
print(data)
```

```python
data_act = [i.text for i in data]
```

```python
print(data_act)
```

```python
newdf= pd.DataFrame(columns = ['Name','Rank','NW','Comp'])
```

```python
newdf
```

```python
data_act[0:4]
print(len(data_act))
```
