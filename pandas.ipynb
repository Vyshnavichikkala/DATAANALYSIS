```python
df3=df2.copy()#copying and printing
df3=df3.dropna()
print(df3)
```

```python
df3.dropna(inplace=True,axis=1)#inplaced says not to reflect the null values changes in the diff
df3
```

```python
df3.dropna(inplace=True,how='all')#drops both rows and columns are not none
df3
```

```python
mv=df3['SEM1'].mean()#it will replace the mean value of mean value
df3=df3.fillna(mv)
print(df3)
```

```python
df3=df3.drop_duplicates()
df3
```

```python
df3.rename(columns={'Grade':'GPA'},inplace=True)
df3.head()
```
